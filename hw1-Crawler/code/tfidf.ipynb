```python
import jieba
import pandas as pd
import ast

N = 5000 #要添加的blog范围

word_dict = {}

for k in range(N):
    print(k)
    df = pd.read_csv(f"./with_local_img/{k+1}.csv",encoding='utf-8')

    paras = ast.literal_eval(df.values[3][1]) #list

    for para in paras:
        if not isinstance(para,list):
            seg_list = jieba.cut_for_search(para)
            for word in seg_list:
                if word not in word_dict:
                    word_dict[word]={'df':1,k+1:1}
                else:
                    if k+1 not in word_dict[word]:
                        word_dict[word][k+1]=1
                        word_dict[word]['df']+=1
                    else:
                        word_dict[word][k+1]+=1
df = pd.DataFrame(word_dict.items())
df.to_csv("word_dict.csv",index=False,encoding='utf-8')
```

```python
import jieba
import pandas as pd
import ast
import math

df = pd.read_csv("word_dict.csv",encoding='utf-8')
N = len(df.values)
tf_idf = {}
for i in range(N):
    word = df.values[i][0]
    tf_idf[word] = {}

for i in range(N):
    print(i)
    word = df.values[i][0]
    word_freqs = ast.literal_eval(df.values[i][1])
    idf = math.log(5000/word_freqs['df'])
    for j in word_freqs:
        if j!='df': tf_idf[word][j] = idf*word_freqs[j]
dfnew = pd.DataFrame(tf_idf.items())
dfnew.to_csv("tf_idf.csv",index=False,encoding='utf-8')
```
